## 4)
The birthday problem is one of the best known paradoxes in probability theory. In its
basic form it is about how likely it is that from a group of n people at least 2 will have a
birthday on the same day. We want to extend this problem and explain it from different
points of view. In both exercises n is limited to the range 2 to 365.

## 4a)
Write a Python function birthday_match(n). This should draw n random birthdays
and check if there is a day that occurs at least 2 times. Plot the probability of a birthday
occurring at least 2 times depending on n.

Only use simulations to determine the probabilities and do not calculate anything analytically!

```python
def birthday_match(n_people_range):
    probabilities = []
    for n in n_people_range:
        matches = 0
        for _ in range(n):
            birthdays = np.random.randint(1, 366, size=n)  # Draw n random birthdays
            unique, counts = np.unique(birthdays, return_counts=True)
            if np.max(counts) >= 2:  # Check if any day occurs at least twice
                matches += 1
        probability = matches / n
        probabilities.append(probability)
    return probabilities


n_people_range = range(2, 366)  # Range of values for n

# Run simulations
probabilities = birthday_match(n_people_range)

# Plot the probability of a birthday occurring at least twice depending on n
plt.plot(n_people_range, probabilities)
plt.xlabel('Number of People (n)')
plt.ylabel('Probability of a Match')
plt.title('Birthday Paradox: Probability of a Match vs Number of People')
# plt.grid(True)
plt.show()
```

## 4b)
Write a Python function your_birthday_match(n). This is to determine the probability
that at least one of n random people will have a birthday exactly with you. Plot
this probability distribution as a function of n and use simulation only!

```python
def your_birthday_match(n_people_range, your_birthday):
    probabilities = []
    for n in n_people_range:
        matches = 0
        for _ in range(n):
            birthdays = np.random.randint(1, 366, size=n) # Draw n random birthdays
            # Check if given birthday matches any of the birthdays
            if your_birthday in birthdays:
                matches += 1
        # Calculate the probability of a match for this value of n
        probability = matches / n
        probabilities.append(probability)
    return probabilities

n_people_range = range(2, 366)  # Range of values for n
# my birthday is the 14th of may -> 31 (January) + 28 (February) + 31 (March) + 30 (April) + 14 (May) = 134
my_birthday = 134

# Run simulations
probabilities = your_birthday_match(n_people_range, my_birthday)

# Plot the probability of given birthday occurring at least twice depending on n
plt.plot(n_people_range, probabilities)
plt.xlabel('Number of People (n)')
plt.ylabel('Probability of a Match')
plt.title('Birthday Paradox: Probability of a Match with Your Birthday vs Number of People')
# plt.grid(True)
plt.show()
```

## 5) Monte-Carlo-Methods II: Approximation of π
The procedure of Monte Carlo simulation is to be clarified by means of the probabilistic
calculation of the circle number π.
For this purpose, random points {(x, y), x ∈ [−1, 1], y ∈ [−1, 1]} are generated and checked
whether they lie within the unit circle. The resulting probability distribution allows the
circle number π to be calculated using the following formula:

$\frac{points within circle}{generated points in square} \hat{=} \frac{r^2π}{(2r)^2} = \frac{π}{4}$

Write a Python function approximate_pi(n). This function should approximate π with
the described Monte-Carlo Method and n generated points.
Also make a plot of the generated points and colorize them, whether they are inside or
outside the circle.

```python
def approximate_pi(n):
    # Generate random points within the square [-1, 1] x [-1, 1]
    x = np.random.uniform(-1, 1, size=n)
    y = np.random.uniform(-1, 1, size=n)

    # Check if each point lies within the unit circle
    distance_from_origin = np.sqrt(x**2 + y**2)
    points_within_circle = distance_from_origin <= 1

    # Approximate pi using the ratio of points within circle to total generated points
    pi_approximation = 4 * np.sum(points_within_circle) / n

    return pi_approximation, x, y, points_within_circle

# Generate points and approximate pi
n = 200000
pi_approximation, x, y, points_within_circle = approximate_pi(n)
print(pi_approximation)

# Plot the generated points and colorize them based on whether they are inside or outside the circle
plt.figure(figsize=(8, 8))
plt.scatter(x, y, c=points_within_circle, cmap='coolwarm', alpha=0.5)
plt.colorbar(label='Inside Circle')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Approximation of π: {pi_approximation:.5f}')
plt.gca().set_aspect('equal', adjustable='box')  # Set aspect ratio to equal
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.grid(True)
plt.show()
```
