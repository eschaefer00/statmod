# Game of coins

```python
# 3a) python function to simulate game of coins
import random 

def simulate_game_of_coins(n: int):
    win_strat_1 = 0
    win_strat_2 = 0 
    
    for _ in range(n):
        coin_tosses = [random.choice(["Heads", "Tails"]) for _ in range(100)]
        you_count = coin_tosses.count("Heads") + coin_tosses.count("Heads")
        friend_count = coin_tosses.count("Heads") + coin_tosses.count("Tails")
        
        if you_count > friend_count:
            win_strat_1 += 1
        elif friend_count > you_count:
            win_strat_2 += 1
    
    return win_strat_1, win_strat_2

strat_1_wins, strat_2_wins = simulate_game_of_coins(1000)

print("Winning Strategy 1 (You):", strat_1_wins, "Winning Strategy 2 (Friend):", strat_2_wins)

if strat_1_wins > strat_2_wins:
    print("Winning Strategy 1 (You) won more games!")
elif strat_2_wins > strat_1_wins:
    print("Winning Strategy 2 (Friend) won more games!")
else:
    print("It's a tie!")

if abs(strat_1_wins - strat_2_wins) > 50:
    print("I'd be surprised if the result was this much skewed, it might be worth double-checking the simulation.")
else:
    print("The result seems reasonable.")
```

```python
import random
import matplotlib.pyplot as plt

def simulate_game(n: int, number_coin_tosses: int):
    win_strat_1 = 0
    win_strat_2 = 0 
    
    for _ in range(n):
        coin_tosses = [random.choice(["Heads", "Tails"]) for _ in range(number_coin_tosses)]
        you_count = coin_tosses.count("Heads") + coin_tosses.count("Heads")
        friend_count = coin_tosses.count("Heads") + coin_tosses.count("Tails")
        
        if you_count > friend_count:
            win_strat_1 += 1
        elif friend_count > you_count:
            win_strat_2 += 1
    
    return win_strat_1, win_strat_2

def simulate_range_of_tosses(n: int, range_of_coin_tosses: int):
    win_strat_1_props = []
    win_strat_2_props = []
    for number in range(2, range_of_coin_tosses + 1):
        strat_1_win_count, strat_2_win_count = simulate_game(n=n, number_coin_tosses=number)
        
        # calculate winning proportion for 1000 simulations of number in range of tosses
        win_strat_1_prop = strat_1_win_count / n
        win_strat_2_prop = strat_2_win_count / n
        
        win_strat_1_props.append(win_strat_1_prop)
        win_strat_2_props.append(win_strat_2_prop)
    
    return win_strat_1_props, win_strat_2_props

# repeat the experiment for all the numbers from 2 to 100 
coin_tosses = 100
strat_1_wins, strat_2_wins = simulate_range_of_tosses(1000, coin_tosses)

# plot the respective proportions
plt.plot(range(2, coin_tosses + 1), strat_1_wins, label="Winning Strategy 1 (You)")
plt.plot(range(2, coin_tosses + 1), strat_2_wins, label="Winnings Strategy 2 (Friend)")
plt.xlabel("Number of Coin Tosses")
plt.ylabel("Win Proportion")
plt.title("Win Proportions of Strategies")
plt.legend()
plt.show()
```
